```python
df1.drop(['world_rank','league_rank', 'club_rank', 'National_rank', 'position_rank',], axis=1, inplace= True)
```

```python
df1.drop(df1.columns[-1], axis=1, inplace= True)
```

```python
df1.drop(df1[df1['apperance'] == 'none'].index.to_list(), axis=0,inplace= True)
```

```python
df1.replace({'apperance_NT':{'none':0},'goals_NT':{'none':0},'assists_NT':{'none':0},'yellow_card_NT':{'none':0},'2nd_ycard_NT':{'none':0},'red_card_NT':{'none':0},'min_NT':{'none':'0"'}}, inplace= True)
```

```python
df1['min'] = df1['min'].apply(lambda val : pd.to_numeric(val[:-1].replace('.','')) if len(val) > 3 else pd.to_numeric(val[:-1]))
```

```python
df1['min_NT'] = df1['min_NT'].apply(lambda val : pd.to_numeric(val[:-1].replace('.','')) if len(val) > 3 else pd.to_numeric(val[:-1]))
```
