we should fill data by scraping from transfermarket site

```python
import requests_html
```

```python
# make url from name of player
# https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query=erling+halland
# extract age for each player
session=requests_html.HTMLSession()
ages=[]
for index,row in df[df["age"].isna()].iterrows():
    url=f'https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={"+".join(row["name"].split(" "))}'
    r=session.get(url)
    age=r.html.find("#yw0 > table > tbody > tr > td:nth-child(4)")[0].text
    ages.append(list([row["name"],age]))
```

```python
# show information about above cell
len(ages),len(ages[0]),ages[0]
```

```python
# save age for each player by name of player
for item in ages:
    df_new.loc[df["name"] == item[0],["age"]]=item[1]
```

fill missing values in positions by scraping position for each pleayer from transfer market

```python
# make url from name of player like which in age cell
# extract position for each player
positions=[]
for index,row in df_v2[df_v2["position"].isna()].iterrows():
    url=f'https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={"+".join(row["name"].split(" "))}'
    r=session.get(url)
    position=r.html.find("#yw0 > table > tbody > tr > td:nth-child(2)")[0].text
    positions.append([row["name"],position])
    
positions
```

```python
# save positions for each player by name of player
for item in positions:
    df_v2.loc[df_v2["name"] == item[0],["position"]]=item[1]
```

fill missing values in Transfer market by scraping this for each pleayer from transfer market site

```python
# make url from name of player like which in age cell
# extract costs (market transfer) for each player
costs=[]
for index,row in df_v2[df_v2["cost"].isna()].iterrows():
    url=f'https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={"+".join(row["name"].split(" "))}'
    r=session.get(url)
    cost=r.html.find("#yw0 > table > tbody > tr > td.rechts.hauptlink")[0].text
    costs.append([row["name"],cost])
# save cost for each player by name of player
for item in costs:
    df_v2.loc[df_v2["name"] == item[0],["cost"]]=item[1]
```

```python
df_v2.isnull().sum()
```

fill missing values in origin_club by scraping origin_club for each pleayer from transfer market site

```python
# make url from name of player like which in age cell and add anoter url for profile each player
# like this https://www.transfermarkt.com/marwan-hamdi/profil/spieler/557938
# extract old_clubs for each player
import re
olds_clubs=[]
for index,row in df_v2[df_v2["origin_club"].isna()].iterrows():
    url=f'https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={"+".join(row["name"].split(" "))}'
    r=session.get(url)
    link=r.html.find("#yw0 > table > tbody > tr > td:nth-child(1) > table")[0].absolute_links
    link_1=0
    for index,j in enumerate(list(link)):
        if re.search("spieler",j) != None:
            link_1=j
    r_profile=session.get(link_1)
    old_club=r_profile.html.find("a.tm-player-transfer-history-grid__club-link")
    olds_clubs.append([row["name"],old_club[0].text])
```

```python
olds_clubs
```

```python
# save origin_club for each player by name of player
for item in olds_clubs:
    df_v2.loc[df_v2["name"] == item[0],["origin_club"]]=item[1]
```

fill missing values in new_club by scraping origin_club for each pleayer from transfer market site

```python
# make url from name of player like which in age cell 
# extract old_clubs for each player
new_clubs=[]
for index,row in df_v2[df_v2["new_club"].isna()].iterrows():
    url=f'https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={"+".join(row["name"].split(" "))}'
    r=session.get(url)
    club=r.html.find('#yw0 > table > tbody > tr > td:nth-child(1) > table')
    new_clubs.append([row["name"],club[0].text.split("\n")[1]])
```

```python
new_clubs
```

```python
# save new_club for each player by name of player
for item in new_clubs:
    df_v2.loc[df_v2["name"] == item[0],["new_club"]]=item[1]
```

league_origin_club,country_origin_club,league_new_club,country_new_club , this columns until contain missing value , we can apply function fillna to fill it by static value = UnKnown
